```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("tsla-data.csv")
df.keys()
```

```python
plt.plot(df["Cash"])
```

```python
b = np.log(df["Cash"])
```

```python
b = b[:-1] # remove NaN
a = np.array([[k,1] for k,_ in enumerate(b)])
```

```python
x, _residuals, _rank, _s = np.linalg.lstsq(a, b, rcond=None)
x, _residuals
```

```python
plt.plot(b)
plt.plot(a @ x)
plt.show()
```
