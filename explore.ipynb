```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("tsla-data.csv")
df.keys()
```

```python
plt.plot(df["Auto"])
```

```python
series = np.log(df["Auto"])
series = series[~np.isnan(series)] # remove NaN
```

```python
quarters = np.array(series.keys())
b = series.values
```

```python
a = np.array([[q, 1] for q in quarters])
```

```python
x, _residuals, _rank, _s = np.linalg.lstsq(a, b, rcond=None)
x, _residuals
```

```python
plt.plot(b)
plt.plot(a @ x)
plt.show()
```

```python
plt.plot(np.exp(a @ x))
```
